```sql
SELECT calculated_race_results.team_name,
       COUNT(1) AS total_races,
       SUM(calculated_race_results.calculated_points) AS total_points,
       AVG(calculated_race_results.calculated_points) as avg_points
  FROM f1_presentation.calculated_race_results
GROUP BY 1
HAVING total_races >= 100
ORDER BY 4 DESC;
```
