```python
dbutils.widgets.text("p_data_source", "")
v_data_source = dbutils.widgets.get("p_data_source")
```

```python
dbutils.widgets.text("p_file_date", "2021-03-21")
v_file_date = dbutils.widgets.get("p_file_date")
```

### **Step 1 - Read the JSON file using the spark dataframe reader**

```python
constructors_schema = "constructorId INT, constructorRef STRING, name STRING, nationality STRING, url STRING"
```

```python
constructors_df = spark.read\
    .schema(constructors_schema)\
        .json(f"{raw_folder_path}/{v_file_date}/constructors.json")
```

### **Step 2 - Drop unwanted columns from the dataframe**

```python
from pyspark.sql.functions import col
```

```python
constructor_dropped_df = constructors_df.drop(col('url'))
```

### **Step 3 - Rename columns and add ingestion date**

```python
from pyspark.sql.functions import lit
```

```python
constructor_final_df = constructor_dropped_df.withColumnRenamed("constructorId", "constructor_id")\
    .withColumnRenamed("constructorRef", "constructor_ref")\
        .withColumn("data-source", lit(v_data_source))\
            .withColumn("file_date", lit(v_file_date))
```

```python
races_with_ingestion_date_df = add_ingestion_date(constructor_final_df)
```

```python
#races_with_ingestion_date_df.write.mode("overwrite").parquet(f"{processed_folder_path}/constructors")
races_with_ingestion_date_df.write.mode("overwrite").format("delta").saveAsTable("f1_processed.constructors")
```
